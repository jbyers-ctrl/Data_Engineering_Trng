## **Syntax:**
```
Syntax: imshow(X, cmap=None, alpha=None, aspect='',interpolation='' )
   - X :- this is input data matrix which is to be displayed
   - cmap :- Colormap we use t dispay the heatmap
   - alpha :- it specifies the opacity or transpiracy of the heatmap.
   - aspect: {'equal', 'auto'} or float,(default: 'equal') The aspect ratio of the plot axes.
   - interpolation: str, (default: 'antialiased') - The interpolation method used.
```

## **Data set information:**

The objective of this example is to visualize the turnover (in dollars) of ice cream sales during the three years using a filled line plot. The chart aims to depict the monthly fluctuations in ice cream sales revenue over the years 2021, 2022, and 2023.



**Example 1: Creating Simple Heat Map for ice cream's sales**

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# Ice cream sales data for the years 2022 and 2023
icecream_sale_year2021 = [30, 38, 25, 18, 6, 12, 7, 14, 17, 20, 27,37]
icecream_sale_year2022 = [2, 7, 14, 17, 20, 27, 30, 38, 25, 18, 6, 1]
icecream_sale_year2023 = [1.5, 3, 10, 13, 22, 36, 30, 33, 24.5, 15, 6.5, 1.2]

# Combine the data into a 2D array (matrix)
icecream_data = np.array([icecream_sale_year2022, icecream_sale_year2023, icecream_sale_year2021])

# Create the heatmap
plt.imshow(icecream_data)


# Add labels and title
plt.xlabel('Month')
plt.ylabel('Year')
plt.title('Ice Cream Sales Heatmap')

# Set x-axis ticks and labels
plt.xticks(np.arange(12), np.arange(1, 13))

# Set y-axis ticks and labels
plt.yticks(np.arange(3), ['2021','2022', '2023'])

# Show the plot
plt.show()
````

**1.2 - Choosing Different Colormaps in Heatmap Using.**
We can choose different colors for Heatmap using the cmap parameter. cmap can help us in making our heatmap more informative.

````python
# Ice cream sales data for the years 2022 and 2023
icecream_sale_year2021 = [30, 38, 25, 18, 6, 12, 7, 14, 17, 20, 27,37]
icecream_sale_year2022 = [2, 7, 14, 17, 20, 27, 30, 38, 25, 18, 6, 1]
icecream_sale_year2023 = [1.5, 3, 10, 13, 22, 36, 30, 33, 24.5, 15, 6.5, 1.2]

# Combine the data into a 2D array (matrix)
icecream_data = np.array([icecream_sale_year2022, icecream_sale_year2023, icecream_sale_year2021])

# Create the heatmap
plt.imshow(icecream_data, cmap='coolwarm')


# Add labels and title
plt.xlabel('Month')
plt.ylabel('Year')
plt.title('Ice Cream Sales Heatmap')

# Set x-axis ticks and labels
plt.xticks(np.arange(12), np.arange(1, 13))

# Set y-axis ticks and labels
plt.yticks(np.arange(3), ['2021','2022', '2023'])

# Show the plot
plt.show()
````

**1.3 - Adding Colorbar to Heatmap Using Matplotlib**

We can add a colorbar to the heatmap using plt.colorbar() function. The colorbar() function shows the weight of color relatively between a certain range.

In the above code:

- We first combine the ice cream sales data for the years 2021, 2022, and 2023 into a 2D array called icecream_data.
- We then create the heatmap using Matplotlib's imshow() function, specifying the colormap (cmap) as 'coolwarm.'
- We add a color bar to the plot using plt.colorbar().
We label the x-axis as "Month," the y-axis as "Year," and provide a title for the plot.
- We set the x-axis ticks to represent the months (1 to 12) and the y-axis ticks to represent the years (2021, 2022, and 2023).
- Finally, we display the plot using plt.show()i.

````python
# Ice cream sales data for the years 2021, 2022 and 2023

icecream_sale_year2021 = [30, 38, 25, 18, 6, 12, 7, 14, 17, 20, 27,37]
icecream_sale_year2022 = [2, 7, 14, 17, 20, 27, 30, 38, 25, 18, 6, 1]
icecream_sale_year2023 = [1.5, 3, 10, 13, 22, 36, 30, 33, 24.5, 15, 6.5, 1.2]

# Combine the data into a 2D array (matrix)
icecream_data = np.array([icecream_sale_year2022, icecream_sale_year2023, icecream_sale_year2021])

# Create the heatmap
plt.imshow(icecream_data, cmap='coolwarm')

# Add color bar
plt.colorbar()

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Year')
plt.title('Ice Cream Sales Heatmap')

# Set x-axis ticks and labels
plt.xticks(np.arange(12), np.arange(1, 13))

# Set y-axis ticks and labels
plt.yticks(np.arange(3), ['2021','2022', '2023'])

# Show the plot
plt.show()
````

## **1.4 - Aspect ratio of the plot**

- In Matplotlib, the aspect parameter controls the aspect ratio of the plot, which is the ratio of the height to the width of the plot. The aspect parameter accepts various values, and 'auto' is one of them.

- When you set `aspect='auto'`, Matplotlib automatically adjusts the aspect ratio of the plot based on the physical aspect ratio of the axes. It tries to maintain the aspect ratio of the data as closely as possible while still fitting within the figure size.

- Setting `aspect='auto'` is useful when you want Matplotlib to determine the aspect ratio dynamically based on the plot's content and the size of the figure, ensuring that the plot looks visually appealing and accurately represents the data.


````python
# Ice cream sales data for the years 2022 and 2023
icecream_sale_year2022 = [2, 7, 14, 17, 20, 27, 30, 38, 25, 18, 6, 1]
icecream_sale_year2023 = [1.5, 3, 10, 13, 22, 36, 30, 33, 24.5, 15, 6.5, 1.2]

# Combine the data into a 2D array (matrix)
icecream_data = np.array([icecream_sale_year2022, icecream_sale_year2023])

# Create the heatmap
plt.imshow(icecream_data, cmap='coolwarm', aspect='auto' )


# Add color bar
plt.colorbar()

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Year')
plt.title('Ice Cream Sales Heatmap')

# Set x-axis ticks and labels
plt.xticks(np.arange(12), np.arange(1, 13))

# Set y-axis ticks and labels
plt.yticks(np.arange(2), ['2022', '2023'])

# Show the plot
plt.show()
````
