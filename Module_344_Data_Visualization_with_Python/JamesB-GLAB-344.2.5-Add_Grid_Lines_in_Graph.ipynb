## **Instructions:**

With Pyplot, you can use the grid() function to add grid lines to the plot.



## **Example 1:  Adding grid lines by using grid() function**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Set plot colors and market
plt.plot(x, y, marker = 'D', linestyle='dotted', color="blue")

# Enabling grid lines:
plt.grid()
plt.show()
```

## **Example 2: Change grid color**
The default color of the grid lines can be overridden by color argument
grid(color="blue").

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Set plot colors and market
plt.plot(x, y, marker = 'D', linestyle='dotted', color="blue")

# Enabling grid lines:
plt.grid(color ="red")
plt.show()
```

## **Example 3: Grid lines above points**

In the previous examples, the grid lines were added behind the points. However, it is possible to add the lines over the data points by setting **`gca().set_axisbelow()`** to **`False`**.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Set plot colors and market
plt.plot(x, y, marker = 'D', linestyle='dotted', color="blue")

# Enabling grid lines:
plt.grid(color="red")
plt.gca().set_axisbelow(False)  # Set grid lines behind other plot elements

plt.show()
```

## **Example 4: Grid lines style**

Grid lines can be customized with the `linestyle` or `ls` argument. In the following example we are setting a dashed line.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Set plot colors and market
plt.plot(x, y, marker = 'D', linestyle='dotted', color="blue")

# Enabling X-axis grid lines:
plt.grid(color ="red",linestyle = "dashed")

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Set plot colors and market
plt.plot(x, y, marker = 'D', linestyle='dotted', color="blue")

# Enabling X-axis grid lines:
plt.grid(color ="m",linestyle = "dashed")

plt.show()
```

## **Example 5: Enable X-axis grid lines only**

The grid lines are added by default both for the X and Y axes, but it is possible to set the grid only for one of them. On the one hand, if you want to add the grid lines for the X axis you can set the axis argument to "x."

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Set plot colors and market
plt.plot(x, y, marker = 'D', linestyle='dotted', color="blue")

# Enabling X-axis grid lines:
plt.grid(color ="red",axis = "x")

plt.show()
```

## **Example 6: Enable Y-axis grid lines only**

On the other hand, if you want to display the grid lines for the Y-axis, just set the axis argument to "y."

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Set plot colors and market
plt.plot(x, y, marker = 'D', linestyle='dotted', color="blue")

# Enabling X-axis grid lines:
plt.grid(color ="red",axis = "y")

plt.show()
```

## **Example 7: Grid Lines transparency**

The **alpha** argument controls the transparency of the grid lines. This argument is very interesting if you want faded grid lines.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Set plot colors and market
plt.plot(x, y, marker = 'D', linestyle='dotted', color="blue")

# Enabling X-axis grid lines:
plt.grid(color ="red",linestyle = "dashed",  alpha = 0.25)


plt.show()

```

## **Example 8: Change background color inside the plot area**

If you want to change the color inside the plot area (i.e., the area between the axes), you can set the facecolor property of the axes object. Here is how you can modify your code to achieve that:


```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Plot with specified marker, linestyle, and color
plt.plot(x, y, marker='D', linestyle='dotted', color="blue")

# Enabling X-axis grid lines:
plt.grid(color="red")
plt.gca().set_axisbelow(True)  # Set grid lines behind other plot elements

# Set the background color inside the plot area
plt.gca().set_facecolor('lightyellow')

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Plot with specified marker, linestyle, and color
plt.plot(x, y, marker='D', linestyle='dotted', color="blue")

# Enabling X-axis grid lines:
plt.grid(color="red", alpha=.3)
plt.gca().set_axisbelow(True)  # Set grid lines behind other plot elements

# Set the background color inside the plot area
plt.gca().set_facecolor('lightgreen')

plt.show()
```
