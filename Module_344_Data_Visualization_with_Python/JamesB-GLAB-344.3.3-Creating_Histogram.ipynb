- The **hist()** function allows creating histograms in Python when using matplotlib.

## **Example 1: Creating a Simple Histogram**

Let's say you have decided to plot the height of your classmate. Now, let's create a very simple histogram with only 10 values:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [160, 163, 163, 170, 164, 170, 180, 172, 173, 190, 175, 180]

plt.hist(data)
plt.show()

```

To make it clearer, we need to specify the values for plt.title(), plt.xlabel(), and plt.ylabel().

```python
data = [160,163,  163, 170, 164, 170, 180, 172, 173, 190,175,180]
plt.hist(data, color="purple")
plt.title("My classmate' height")
plt.ylabel("Number of people")
plt.xlabel("Height in cm")
plt.show()
```

## **Horizontal Histogram**
Setting the orientation argument to "horizontal," you can flip the axes to create a horizontal histogram.

```python
data = [160,163,  163, 170, 164, 170, 180, 172, 173, 190,175,180]
plt.hist(data, color="purple", orientation = "horizontal")
plt.title("My classmate' height")
plt.xlabel("Number of people")
plt.ylabel("Height in cm")
plt.show()

```

**Changing bins:**
Let's say we want to divide our data values into 3 groups: from **160 to 170 cm, from 170 to 180,** and **from 180 to 190**. To do that, we need to pass a list of these values to the bins argument:

**How to use bin:** The bins argument can take not only a list but also an int. A list defines bin edges. An int defines the number of equal-width bins. So, if we want to have four bins in our histogram, we simply write bins=4. There is also a third option – use a str as bins value. It has to be a name of one of the binning functions supported by numpy, such as 'rice,' 'scott,' 'sqrt,' and so on.

```python
data = [160,163,  163, 170, 164, 170, 180, 172, 173, 190,175,180]
plt.hist(data, bins=3, color="purple")
plt.title("My classmate' height")
plt.ylabel("Number of people")
plt.xlabel("Height in cm")
plt.show()
```

## **Cumulative Histogram**

The hist() function also allows creating cumulative histograms, just setting **cumulative = True** as argument.

```python
data = [160,163,  163, 170, 164, 170, 180, 172, 173, 190,175,180]
plt.hist(data, color="purple",cumulative = True)
plt.title("My classmate' height")
plt.ylabel("Number of people")
plt.xlabel("Height in cm")
plt.show()
```

## **Border Color**

By default, the border color of the bins is the same as the fill color, but you can use the edgecolor argument to customize its color. The following will create a “classic” histogram where the bins are independent.

```python
data = [160,163,  163, 170, 164, 170, 180, 172, 173, 190,175,180]
plt.hist(data, color="purple", edgecolor = "white")
plt.title("My classmate' height")
plt.ylabel("Number of people")
plt.xlabel("Height in cm")
plt.show()

```

# **Histogram by Group**

You can also create a histogram with several groups with the function, just passing a list of data sets, as in the example below. By default, the bins of the different groups will be displayed side by side.

**Note that when creating a histogram with multiple data sets, the datasets does not need to be of the same length.**

```python
# Seed for reproducibility
np.random.seed(4)

# Data simulation
x1 = np.random.normal(0, 1, 100)
x2 = np.random.normal(0, 1, 100)
labelName = ['DataSetOne', 'DataSetTwo']

# Histogram
fig, ax = plt.subplots()
ax.hist([x1, x2], label = labelName)
plt.legend()
plt.show()



```

## **Barstacked**
By default bars are placed side-by-side.The alternative way is to stack the values on top of each other. You can do that by setting the stacked argument to True. We also add edgecolor for better readability:



```python
# Seed for reproducibility
np.random.seed(4)

# Data simulation
x1 = np.random.normal(0, 1, 100)
x2 = np.random.normal(0, 1, 100)
labelName = ['DataSetOne', 'DataSetTwo']

# Histogram
fig, ax = plt.subplots()
ax.hist([x1, x2], label = labelName, stacked=True)
plt.legend()
plt.show()
```

## **Color by Group**

If your data contains several groups you can pass a list of colors to the color argument to customize the colors of the groups.

```python
# Seed for reproducibility
np.random.seed(4)

# Data simulation
x1 = np.random.normal(0, 1, 100)
x2 = np.random.normal(0, 1, 100)
labelName = ['DataSetOne', 'DataSetTwo']

# Histogram
fig, ax = plt.subplots()
ax.hist([x1, x2], label = labelName, stacked=True, color = ["lightsalmon", "mediumaquamarine"])
plt.legend()
plt.show()
```

## **Step Method**

The "step" method generates an unfilled line plot; this is, it shows, the outer border of the bins.

```python
# Seed for reproducibility
np.random.seed(4)

# Data simulation
x1 = np.random.normal(0, 1, 100)
x2 = np.random.normal(0, 1, 100)
labelName = ['DataSetOne', 'DataSetTwo']

# Histogram
fig, ax = plt.subplots()
ax.hist([x1, x2], label = labelName, histtype = "step", stacked=True)
plt.legend()
plt.show()
```
