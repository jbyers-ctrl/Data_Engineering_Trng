## **Introduction:**

Here is an explanation of how to use Matplotlib's Pyplot subplots:

**Creating Subplots:**
You can create subplots using the plt.subplots() function. This function returns a figure object and a multidimensional array of axes objects. The syntax is as follows:


```
fig, axs = plt.subplots(nrows, ncols)


```
Where nrows and ncols are the number of rows and columns in the subplot grid, respectively:


## **Instructions:**


### **Example 1**

Here's a simple example for demonstrating how to create and work with subplots:



````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# the figure has 1 row, 2 columns, and this plot is the first plot
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("plot 1")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# the figure has 1 row, 2 columns, and this plot is the second plot
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title("plot 2")

plt.show()
````

## **Example 2**

This example creates a figure with two subplots arranged vertically. It plots y1 on the first subplot and y2 on the second subplot.

````python
# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 9, 16, 25]

# Create subplots using Object Oriented Approach
fig, axs = plt.subplots(2, 1)

# Plot data on subplots
axs[0].plot(x, y1)
axs[0].set_title('Plot 1')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')

axs[1].plot(x, y2)
axs[1].set_title('Plot 2')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis')

plt.tight_layout()  # plt.tight_layout() function used to automatically adjust the layout of subplots within a figure.
plt.show()
````

- In Matplotlib, **fig** refers to the "figure" object, which serves as the top-level container for all elements of a plot. It represents the entire window or page where your plots are drawn. Understanding fig is crucial for customizing and organizing your visualizations effectively.

- In Python's Matplotlib library, **plt.tight_layout()** is a function used to automatically adjust the layout of subplots within a figure. Here is a breakdown of its functionality:
-- Purpose: It addresses the issue of overlapping elements in figures containing multiple subplots such as titles, labels, and tick marks.
-- An alternative to **plt.tight_layout()** is** plt.subplots_adjust()**, which provides more granular control over subplot spacing but requires manual adjustments.
-- Overall, plt.tight_layout() is a helpful tool for quickly improving the appearance of figures with multiple subplots in Matplotlib.



## **Example 3: Creating a 2x2 Subplots Grid:**

This example creates a 2x2 grid of subplots and plots different datasets in each subplot.

````python
# creates an array of 100 evenly spaced numbers between 0 and 2 * np.pi
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

fig, axes = plt.subplots(2, 2)

axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sin Function')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cos Function')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Tan Function')

axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Exponential Function')

# add a title to the entire figure with the suptitle() function
plt.suptitle("Trigonometry")
plt.tight_layout()
plt.show()
````

## **Example 4: Creating a Vertical Stack of Subplots:**
This example creates a vertical stack of three subplots.

````python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig, axes = plt.subplots(3, 1)

axes[0].plot(x, y1)
axes[0].set_title('Sin Function')

axes[1].plot(x, y2)
axes[1].set_title('Cos Function')

axes[2].plot(x, y3)
axes[2].set_title('Tan Function')

plt.suptitle("Trig")
plt.tight_layout()
plt.show()
````

## **Example 5: Sharing Axes between Subplots:**


This is an advanced example. In this example, a 2x2 grid of subplots is created with a shared x-axis and y-axis.

````python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        if i == 0:
            axes[i, j].plot(x, y1)
            axes[i, j].set_title('Sin Function')
        else:
            axes[i, j].plot(x, y2)
            axes[i, j].set_title('Cos Function')

plt.tight_layout()
plt.show()
````
