## **Objective:**
In this lab, we will demonstrate how to create multiple lines in Matplotlib Pyplot.

By the end of this lab, learners will be able to create multiple line graphs in a single figure.

Note:
- If no color is specified, Matplotlib will automatically cycle through a set of default colors for multiple lines.

- You can adjust the line style using the linestyle keyword.

## **Example 1: Creating mutiple lines for World Population Graph**

```python
import matplotlib.pyplot as plt
```

```python
# Declaring Years and populations Dataset
years=[1950,1955,1960,1965,1970,1980,1985,1990,1995,2000,2005,2010,2015]
pops=[2.5,2.7,3.0,3.3,3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3]
death=[1.2,1.1,1.2,2.1,2.0,2.3,1.8,1.9,2.6,1.6,2.4,2.4,4.0]

# Plotting the data
plt.plot(years,pops, color=('red'))
plt.plot(years,death, color=('green'))

plt.ylabel("Population in Billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
plt.show()
```

## **Example 2: Let's change the line style**
In this example we will change the line style from solid to dashed.

```python
years=[1950,1955,1960,1965,1970,1980,1985,1990,1995,2000,2005,2010,2015]
pops=[2.5,2.7,3.0,3.3,3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3]
death=[1.2,1.1,1.2,2.1,2.0,2.3,1.8,1.9,2.6,1.6,2.4,2.4,4.0]
plt.plot(years,pops,'--', color=('red'))
plt.plot(years,death, '-.', color=('green'))
plt.ylabel("Population in Billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
plt.show()
```

## **Example 3: Let's add the markets for each line**

In this example, we will add the marker for the data points of each line.

```python
years=[1950,1955,1960,1965,1970,1980,1985,1990,1995,2000,2005,2010,2015]
pops=[2.5,2.7,3.0,3.3,3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3]
death=[1.2,1.1,1.2,2.1,2.0,2.3,1.8,1.9,2.6,1.6,2.4,2.4,4.0]
plt.plot(years,pops,'--', color=('red'), marker="*")
plt.plot(years,death,'-.', color=('green'), marker="*")
plt.ylabel("Population in Billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
plt.show()
```

## **Example 4: Let's add legend for each line**

When multiple lines are displayed within a single axes, it is often helpful to create a plot legend that labels each line type. Matplotlib provides a built-in method for quickly creating such a legend using the `plt.legend()` function. You can specify labels for each line by using the label keyword argument in the `plot()` function.  

```python
years=[1950,1955,1960,1965,1970,1980,1985,1990,1995,2000,2005,2010,2015]
pops=[2.5,2.7,3.0,3.3,3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3]
death=[1.2,1.1,1.2,2.1,2.0,2.3,1.8,1.9,2.6,1.6,2.4,2.4,4.0]
plt.plot(years,pops,'--', color=('red'), marker="*", label="population" )
plt.plot(years,death,'-.', color=('green'), marker="*", label="death")
plt.ylabel("Population in Billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
plt.legend()
plt.show()
```

## **Example 5: Lets add one more line graph for job rate.**

```python
years = [1950, 1955, 1960, 1965, 1970, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]
pops = [2.5, 2.7, 3.0, 3.3, 3.6, 4.0, 4.4, 4.8, 5.3, 5.7, 6.1, 6.5, 7.3]
death = [1.2, 1.1, 1.2, 2.1, 2.0, 2.3, 1.8, 1.9, 2.6, 1.6, 2.4, 2.4, 4.0]
job_rate = [0.5, 0.6, 0.8, 0.9, 1.0, 1.2, 1.5, 1.7, 2.0, 2.2, 2.5, 2.7, 2.9]
plt.figure(figsize=(10, 10))
plt.plot(years, pops, '--', color='red', marker="*", label="population")
plt.plot(years, death, '-.', color='green', marker="*", label="death")
plt.plot(years, job_rate, ':', color='blue', marker="*", label="job rate")  # Plot job rate data
plt.ylabel("Population in Billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
plt.legend()

plt.show()
```
