```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Create sample data
expense_category = ['Groceries', 'Utilities', 'Transportation', 'Dining Out', 'Entertainment']
amount_spent = [500, 300, 200, 400, 250]

# Create colors
colors = ['lightgreen', 'lightblue', 'magenta', 'lightpink', 'brown']

# Create horizontal bar
plt.barh(expense_category, amount_spent, color=colors, height=0.5)

# Add labels to x & y axis
plt.xlabel("Amount Spent")
plt.ylabel("Expense Categories")

# Add title to chart
plt.title("Monthly Expenses Distribution")

# Display chart
plt.show()
```
