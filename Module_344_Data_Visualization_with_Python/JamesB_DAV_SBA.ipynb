```python
# Imported Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data from CSV file
df = pd.read_csv("./Data/woocommerce-product-export.csv")
df
```

```python
# Show concise summary of columns
df.info()
```

```python
# Show summary of statistics
df.describe()
```

```python
# Print first five rows
print("-"*40, "First five rows", "-"*40)
print(df.head())

# Print last five rows
print("-"*40, "Last five rows", "-"*40)
print(df.tail())
```

```python
# Print "total_profit" & "month_number" columns only
df[['total_profit', 'month_number']]
```

```python
# Read total profit of all months & show with a bar plot
sns.barplot(x='month_number', y='total_profit', data=df)
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.title('Company profit per month')
plt.show()
```

```python
# Read total profit of all month & show with a line plot
sns.lineplot(x='month_number', y='total_profit', data=df, linestyle='dashed', color='red', linewidth=3, marker='o', markerfacecolor='black')
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.title('Company profit per month')
plt.legend(title='Profit data of last year', loc='lower right')
plt.show()
```

```python
# Print all product sales data & show with a multi-line plot
plt.figure(figsize=(8, 5))
plt.plot(df['month_number'], df['facecream'], marker='o', label='Face cream Sales Data')
plt.plot(df['month_number'], df['facewash'], marker='o', label='Face Wash Sales Data')
plt.plot(df['month_number'], df['toothpaste'], marker='o', label='Toothpaste Sales Data')
plt.plot(df['month_number'], df['bathingsoap'], marker='o', label='Bathingsoap Sales Data')
plt.plot(df['month_number'], df['shampoo'], marker='o', label='Shampoo Sales Data')
plt.plot(df['month_number'], df['moisturizer'], marker='o', label='Moisturizer Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sale units in number')
plt.title('Sales Data')
plt.legend()
plt.show()
```

```python
# Read bathingsoap sales data for each month & show with a scatter plot
plt.scatter(df['month_number'], df['bathingsoap'], marker='o', label='bathing soap Sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.title('bathingsoap Sales data')
plt.grid(linestyle='dashed')
plt.legend()
plt.show()
```

```python
# Dataset
date = ["25/12", "26/12", "27/12"]
temp = [8.5, 10.5, 6.8]

# Convert dataset to a pandas dataframe
date_temp_df = pd.DataFrame(zip(date, temp), columns=['Date', 'Temperature'])

# Show line chart of date versus temperature
plt.plot(date_temp_df['Date'], date_temp_df['Temperature'])
plt.xlabel('Date')
plt.ylabel('temperature')
plt.title('Date-wise Temperature')
plt.show()
```
