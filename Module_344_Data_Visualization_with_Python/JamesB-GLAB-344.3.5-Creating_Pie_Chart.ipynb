## **Syntax:**

The **plt.pie(data)** function allows creating pie chart in Python when using matplotlib.

Where data is an array of data values that you want to plot, there are many other optional parameters that can pass to plt.pie(). Some of them include:
- Explode allows separating slices of a pie chart.
- Labels is a list of strings that sets the label of each corresponding slice
- Labeldistance determines the radial distance at which pie labels are drawn (default is 1.1).
- Shadow creates the shadow of a slice.
- Startangle lets you choose the starting angle of a plot.
- Wedgeprops lets you tune various slice parameters.

Note: The size of each wedge is determined by comparing the value with all of the other values, by using this formula:

The value divided by the sum of all values: x/sum(x)

## **Example 1: Create Simple Pie Chart**
Let's create a pie chart representing the distribution of different fruits in a store.


```python
import matplotlib.pyplot as plt
```

```python
labelsName = 'apple', 'banana', 'cherry', 'durian', 'elderberries', 'figs', 'grapes'
sizes = [32, 20, 15, 10, 10, 8, 5]

plt.pie(sizes, labels=labelsName)
plt.title('Fruits in the store')
plt.tight_layout()
plt.legend(labelsName)
plt.show()
```

## **Example 2: Start Angle**
- By default, pie charts are plotted with the first slice starting from the positive x-axis and proceeding counterclockwise.
- We can change the start angle by specifying a **startangle** parameter.
- The **startangle** parameter is defined in degrees, with the default angle being 0.

In this example, we will demonstrate both clockwise and counter-clockwise pie charts using the object-oriented approach by creating subplots.

img_matplotlib_pie_angles.png

```python

labelsName = 'apple', 'banana', 'cherry', 'durian', 'elderberries', 'figs', 'grapes'
sizes = [32, 20, 15, 10, 10, 8, 5]

fig, axes = plt.subplots(1, 2, figsize=(20, 6))
ax1, ax2 = axes

ax1.pie(sizes, labels=labelsName,startangle=90)
ax2.pie(sizes, labels=labelsName,  startangle=90, counterclock=False)
ax1.set_title('Starting the plot at 90')
ax2.set_title('Plotting clockwise')
# plt.title('Fruits in the store')
plt.tight_layout()
plt.legend(labelsName)
plt.show()
```

## **Example 3: Explode**
The **explode** parameter in pie charts separates a slice from the main pie. To highlight some of the fractions, we need to pass a list of values to the explode parameter. The values determine how far we want to separate this category from the main pie.

```python
labelsName = 'apple', 'banana', 'cherry', 'durian', 'elderberries', 'figs', 'grapes'
sizes = [32, 20, 15, 10, 10, 8, 5]
explode = [0.08, 0.0, 0.0, 0.0, 0.08, 0.0, 0.0]
plt.pie(sizes, labels=labelsName, explode=explode)
plt.title('Fruits in the store')
plt.tight_layout()
plt.legend(labelsName)
plt.show()
```

## **Example 4:Display the value**
The **autopct** parameter displays the percent value on the slices. If **autopct** is **%.2f**, then for each pie slice, the format string is %.2f where:
- **%** is a special character that tells when to type the value.
- **f** sets the result to be a floating-point type.
- **.2** sets a limit to only 2 digits after the point.

```python
labelsName = 'apple', 'banana', 'cherry', 'durian', 'elderberries', 'figs', 'grapes'
sizes = [32, 20, 15, 10, 10, 8, 5]


explode = [0.08, 0.0, 0.0, 0.0, 0.08, 0.0, 0.0]

plt.pie(sizes,  labels=labelsName,    explode=explode,
         autopct='%.2f%%'   # here we also add the % sign
        )
plt.title('Fruits in the store')

plt.tight_layout()
plt.legend(labelsName)
plt.show()
```

# **Example 5: Add the frame of the plot**

You might have noticed that the default pie does not display the typical frame of the charts created with matplotlib. In case you want to add it, you can set **frame = True.**

```python
labelsName = 'apple', 'banana', 'cherry', 'durian', 'elderberries', 'figs', 'grapes'
sizes = [32, 20, 15, 10, 10, 8, 5]


explode = [0.08, 0.0, 0.0, 0.0, 0.08, 0.0, 0.0]

plt.pie(sizes,  labels=labelsName,  explode=explode, frame="True",
         autopct='%.2f%%'   # here we also add the % sign
        )
plt.title('Fruits in the store')

plt.tight_layout()
plt.legend(labelsName)
plt.show()
```

## **Example 6: Group labels distance to the origin**

In addition, you can also modify the default distance (1.1 for a slice of radius 1) of the group labels with labeldistance.

```python
labelsName = 'apple', 'banana', 'cherry', 'durian', 'elderberries', 'figs', 'grapes'
sizes = [32, 20, 15, 10, 10, 8, 5]


explode = [0.08, 0.0, 0.0, 0.0, 0.08, 0.0, 0.0]

plt.pie(sizes,  labels=labelsName,  explode=explode, frame="True", labeldistance = 0.8,
         autopct='%.2f%%'   # here we also add the % sign
        )
plt.title('Fruits in the store')

plt.tight_layout()
plt.legend(labelsName)
plt.show()
```

## **Example 6: Customizing Colors:**
- You can set the color of each wedge with the colors parameter.

- The colors parameter, if specified, must be an array with one value for each wedge:

- You can use Hexadecimal color values, any of the 140 supported color names, or one of these shortcuts:

'r' - Red
'g' - Green
'b' - Blue
'c' - Cyan
'm' - Magenta
'y' - Yellow
'k' - Black
'w' - White

```python
labelsName = 'apple', 'banana', 'cherry', 'durian', 'elderberries', 'figs', 'grapes'
sizes = [32, 20, 15, 10, 10, 8, 5]

mycolors = colors = ["#B9DDF1", "#9FCAE6", "#73A4CA", "#497AA7", "#2E5B88"]
explode = [0.08, 0.0, 0.0, 0.0, 0.08, 0.0, 0.0]

plt.pie(sizes,
        labels=labelsName,
        explode=explode,
        colors=mycolors)
plt.title('Fruits in the store')

plt.tight_layout()
plt.legend(labelsName)
plt.show()
```
