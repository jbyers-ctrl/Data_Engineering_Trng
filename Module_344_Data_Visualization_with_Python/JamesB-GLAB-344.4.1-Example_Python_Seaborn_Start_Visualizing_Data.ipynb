### **Example 1 - Seaborn Scatter Plots:**

Scatter plots are used to visualize the relationship between two continuous variables. Each point on the plot represents a single data point, and the position of the point on the x and y-axis represents the values of the two variables.

**1.1** - The plot can be customized with different colors and markers to help distinguish different groups of data points. In Seaborn, scatter plots can be created using the scatterplot() function.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(10)
```

- The above simple plot can be improved by customizing the `hue` and `size` parameters of the plot.

 - The `hue` parameter is used to add a categorical variable to the plot by coloring the points based on different categories. In this example, the points in the scatter plot will be colored differently based on the 'sex' variable, with each category ('Male' and 'Female') represented by a different color. This allows for visual comparison of the relationship between 'total_bill' and 'tip' across different categories of the 'sex' variable.

 **1.2 - Here is an example**

```python
# customize the scatter plot

sns.scatterplot(x="total_bill", y="tip", hue="sex", style="sex", size="size", sizes=(50, 200), data=tips)

# Add a title and labels to the plot using Matplotlib

plt.xlabel("Total Bill")

plt.ylabel("Tip")

plt.title("Relationship between Total Bill and Tip")


# display the plot

plt.show()
```

In this example, we have used the `seaborn` library for simple scatter plot and have used `matplotlib` for further customizing the scatter plot.

### **Example 2 - Seaborn Line Plots:**
**2.1** - Line plots are used to visualize trends in data over time or other continuous variables. In a line plot, each data point is connected by a line, creating a smooth curve.
In Seaborn, line plots can be created using the lineplot() function.

```python
fmri = sns.load_dataset("fmri")
fmri.head(5)
```

```python
sns.lineplot(x="timepoint", y="signal", sizes=(50, 200), data=fmri)
```

**2.2** - We can very easily customize this by using `event` and `region` columns from the dataset.

```python
fmri = sns.load_dataset("fmri")
# customize the line plot

sns.lineplot(x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False, sizes=(70, 200), data=fmri)


# Add a title and labels to the plot using Matplotlib

plt.xlabel("Timepoint")

plt.ylabel("Signal Intensity")

plt.title("Changes in Signal Intensity over Time")


# display the plot

plt.show()
```

### **2.3 - Vertical Lines with axvline() Functions**

If you want to add a vertical line to a plot made with Seaborn, you can use the axvline function and set where you want to display the vertical line with the `x` argument.

```python
ax = sns.lineplot(x="timepoint", y="signal", color = "limegreen", alpha = 0.1, data=fmri)
ax.axvline(x = 5)    # Line on x = 5


plt.xlabel("Timepoint")

plt.ylabel("Signal Intensity")

plt.title("Changes in Signal Intensity over Time")
```

## **Horizontal lines with axhline() functions**
Similarly, the axhline() function allows adding horizontal lines to the Seaborn plots, but now you need to set the value with the `y`argument.



```python
ax = sns.lineplot(x="timepoint", y="signal", color = "limegreen", alpha = 0.1, data=fmri)
ax.axvline(x = 5)    # Line on x = 5

ax.axhline(y = 0.17,    linestyle = "dashed")

# Add a title and labels to the plot using Matplotlib

plt.xlabel("Timepoint")

plt.ylabel("Signal Intensity")

plt.title("Changes in Signal Intensity over Time")
```

Again, we have used the `Seaborn` library to do a simple line plot and have used the `matplotlib` library to customize and improve the simple line plot.

### **Example 3 - Seaborn bar plots**
Bar plots are used to visualize the relationship between a categorical variable and a continuous variable. In a bar plot, each bar represents the mean or median (or any aggregation) of the continuous variable for each category.

**3.1** - In Seaborn, bar plots can be created using the barplot() function.


```python
titanic = sns.load_dataset("titanic")
titanic.head(5)
```

```python
sns.barplot(x="class", y="fare", data=titanic)
```

**3.2** - Let’s customize the above bar plot, showing the fares for different classes, with bars differentiated by gender (male and female).

The hue parameter allows differentiation of the bars based on the "sex" column from the dataset. It separates the data into categories based on the values in the "sex" column, enabling a comparison between males and females within each class.

```python
titanic = sns.load_dataset("titanic")

# customize the bar plot

sns.barplot(x="class", y="fare", hue="sex", data=titanic)

# Add a title and labels to the plot using Matplotlib

plt.xlabel("Class")

plt.ylabel("Fare")

plt.title("Average Fare by Class and Gender on the Titanic")


# display the plot

plt.show()
```

The `hue` parameter is used to add another categorical variable to the plot by changing the color of the bars based on the categories of this variable. In this case, the bars representing the 'fare' for each 'class' on the x-axis are further divided into different colors based on the 'sex' of the passengers. This helps in visualizing the distribution of fares within each class while also considering the gender of the passengers.

### **Example 4 - Seaborn histograms:**

Histograms visualize the distribution of a continuous variable. In a histogram, the data is divided into bins and the height of each bin represents the frequency or count of data points within that bin.

**4.1** - In Seaborn, histograms can be created using the histplot() function.


```python
iris = sns.load_dataset("iris")
iris.head(10)
```

```python
sns.histplot(x="petal_length", data=iris)
```

**4.2** - Customizing a histogram,

Lets add `bin`, `kde` and `color` parameters:

**bins**: Defines the number of bins or intervals into which the data range will be divided. Here, it is set to 20 bins.

**kde**: Determines whether to plot a kernel density estimate (KDE) curve along with the histogram. It is set to True, indicating that the KDE curve will be included.


```python
iris = sns.load_dataset("iris")

# customize the histogram

sns.histplot(data=iris, x="petal_length", bins=20, kde=True, color="green")

# Add a title and labels to the plot using Matplotlib

plt.xlabel("Petal Length (cm)")

plt.ylabel("Frequency")

plt.title("Distribution of Petal Lengths in Iris Flowers")

# display the plot

plt.show()
```

### **Exampple 5 - Seaborn density plots:**
Density plots, also known as kernel density plots, are a type of data visualization that display the distribution of a continuous variable. They are similar to histograms, but instead of representing the data as bars, density plots use a smooth curve to estimate the density of the data.

**5.1** - In Seaborn, density plots can be created using the **kdeplot()** function.


```python
# Create a density plot of the "total_bill" column from the "tips" dataset

tips = sns.load_dataset("tips")

sns.kdeplot(data=tips, x="total_bill")
```

**5.2** - Let’s improve the plot by customizing it.

- We can use the `hue` parameter to differentiate between "lunch" and "dinner" meal times.

- We use the `fill` parameter to fill the area under the curve.

- We adjust the `alpha` and `linewidth` parameters to make the plot more visually appealing.

```python
# Load the "tips" dataset from Seaborn

tips = sns.load_dataset("tips")

# Create a density plot of the "total_bill" column from the "tips" dataset

# We use the "hue" parameter to differentiate between "lunch" and "dinner" meal times

# We use the "fill" parameter to fill the area under the curve

# We adjust the "alpha" and "linewidth" parameters to make the plot more visually appealing

sns.kdeplot(data=tips, x="total_bill", hue="time", fill=True, alpha=0.5, linewidth=1.5)


# Add a title and labels to the plot using Matplotlib

plt.title("Density Plot of Total Bill by Meal Time")

plt.xlabel("Total Bill ($)")

plt.ylabel("Density")




# Show the plot

plt.show()
```

**5.3 - Lets add Horizontal lines and verticle line.**

We can use `axhline()` and `axvline()` functions.

```python
# Load the "tips" dataset from Seaborn

tips = sns.load_dataset("tips")

# Create a density plot of the "total_bill" column from the "tips" dataset

# We use the "hue" parameter to differentiate between "lunch" and "dinner" meal times

# We use the "fill" parameter to fill the area under the curve

# We adjust the "alpha" and "linewidth" parameters to make the plot more visually appealing

p = sns.kdeplot(data=tips, x="total_bill", hue="time", fill=True, alpha=0.5, linewidth=1.5)

p.axvline(x = 17 )
p.axhline(y = 0.036)


# Add a title and labels to the plot using Matplotlib

plt.title("Density Plot of Total Bill by Meal Time")

plt.xlabel("Total Bill ($)")

plt.ylabel("Density")


# Show the plot

plt.show()
```

### **Example 6 - Seaborn Box Plots**
The Box plots are a type of visualization that shows the distribution of a dataset. They are commonly used to compare the distribution of one or more variables across different categories.

**6.1** - In Seaborn, boxplot plots can be created using the boxplot() function.




```python
# load the tips dataset from Seaborn
tips = sns.load_dataset("tips")

sns.boxplot(x="day", y="total_bill", data=tips)
```

**6.2** - Customize the box plot by including `time` column from the dataset, and by adding `hue`, `palette` `linewidth` and `fliersize` parameters.

- We can use the `hue` parameter to differentiate between "lunch" and "dinner" meal times.

- We can customize the color scheme by using the `palette` parameter.

- We can adjust the `linewidth` and `fliersize` parameters to make the plot more visually appealing.

```python
# load the tips dataset from Seaborn

tips = sns.load_dataset("tips")

# create a box plot of total bill by day and meal time, using the "hue" parameter to differentiate between lunch and dinner

# customize the color scheme using the "palette" parameter

# adjust the linewidth and fliersize parameters to make the plot more visually appealing

p = sns.boxplot(x="day", y="total_bill", hue="time", data=tips, palette="Set3", linewidth=1.5, fliersize=4)


# add a title, xlabel, and ylabel to the plot using Matplotlib functions

plt.title("Box Plot of Total Bill by Day and Meal Time")

plt.xlabel("Day of the Week")

plt.ylabel("Total Bill ($)")


# display the plot

plt.show()
```

### **Example 7 - Seaborn Heatmaps**
A heatmap is a graphical representation of data that uses colors to depict the value of a variable in a two-dimensional space. Heatmaps are commonly used to visualize the correlation between different variables in a dataset.

**7.1** - The heatmap() function allows creating heat maps or correlations plots in Python with Seaborn. You can pass a 2D data set or a pandas dataframe. In case you provide a pandas data frame the row and column names will be used to label the plot axes.


```python
# Load the dataset

tips = sns.load_dataset('tips')

# Create a heatmap of the correlation between variables

corr = tips.corr()

sns.heatmap(corr)

# Show the plot

#plt.show()
```

**Explanation**:

- **tips:** This is the DataFrame containing the dataset loaded using sns.load_dataset('tips'). It contains various columns representing different aspects of tips data, such as total bill amount, tip amount, sex, smoker status, day of the week, etc.

- **.corr():** This is a method provided by Pandas DataFrames to calculate the correlation between numerical columns. When applied to a DataFrame, it returns another DataFrame containing the correlation coefficients between all pairs of numerical columns. The correlation coefficient measures the strength and direction of a linear relationship between two variables. It ranges from -1 to 1, where 1 indicates a perfect positive correlation, -1 indicates a perfect negative correlation, and 0 indicates no correlation.

- **corr:** This is the variable where the resulting correlation DataFrame is stored. Each cell in the corr DataFrame represents the correlation coefficient between the variables corresponding to the row and column indices.

- So,** corr = tips.corr()** computes the correlation matrix for the numerical variables in the tips DataFrame and stores it in the variable corr. This correlation matrix is then used to create the heatmap using sns.heatmap(corr), which visualizes the correlation coefficients between different pairs of variables.

### **7.3 - Another example of a heatmap using the `flights` dataset.**

In this example, we will use the `flights` dataset from the `seaborn` library.

```python
# Load the dataset
flights = sns.load_dataset('flights')


# Pivot the data

flights = flights.pivot('month', 'year', 'passengers')
flights.head(5)
```

```python
# Create a heatmap

sns.heatmap(flights, cmap='Blues', annot=True, fmt='d')


# Set the title and axis labels

plt.title('Passengers per month')

plt.xlabel('Year')

plt.ylabel('Month')


# Show the plot

plt.show()
```

In this example, we are using the `flights` dataset from the `Seaborn` library. We pivot the data to make it suitable for heatmap representation using the .pivot() method. Then we create a heatmap using the sns.heatmap() function and pass the pivoted flights variable as the argument.



### **Example 8 - Seaborn pair plots**

**8.1 -** Pair plots are a type of visualization in which multiple pairwise scatter plots are displayed in a matrix format. Each scatter plot shows the relationship between two variables, while the diagonal plots show the distribution of the individual variables.


```python
# Load iris dataset

iris = sns.load_dataset("iris")


# Create pair plot

sns.pairplot(data=iris)


# Show plot

plt.show()
```

**8.2 -** We can customize this plot by using `hue` and `diag_kind` parameter.

```python
# Load iris dataset

iris = sns.load_dataset("iris")


# Create pair plot with custom settings

sns.pairplot(data=iris, hue="species", diag_kind="kde", palette="husl")


# Set title

plt.title("Iris Dataset Pair Plot")

# Show plot

plt.show()
```
