## **Lab objective:**

In this lab, we will demonstrate how to control the properties of the figure by using `plt.figure()`.



## **Introduction:**
**plt.figure()** allows you to explicitly control the properties of the figure before plotting anything on it. We can also specify various parameters when calling plt.figure() to customize the figure such as its size, aspect ratio, and background color.



In this example, **plt.figure()** creates a new figure object, which is then used to plot some data. If you do not call **plt.figure()**, Matplotlib will automatically create a new figure when you first plot something. However, using **plt.figure()** allows you to explicitly control the properties of the figure before plotting anything on it.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

# Create a new figure with default parameters
fig = plt.figure()

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Plot with specified marker, linestyle, and color
plt.plot(x, y, marker='D', linestyle='dotted', color="blue")

# Enabling X-axis grid lines:
plt.grid(color="red")
plt.savefig("gfgfg")
```

You can also specify various parameters when calling **plt.figure()** to customize the figure such as the figure size, aspect ratio, background color, etc. For example:

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

#creates a new figure with a custom size of 9x10 inches.
fig = plt.figure(figsize=(9, 10))

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Plot with specified marker, linestyle, and color
plt.plot(x, y, marker='D', linestyle='dotted', color="blue")

# Enabling X-axis grid lines:
plt.grid(color="red")

plt.show()
```

The **dpi** can be passed as an argument to the figure() function. We can control the width of objects (axes, fonts, lines, markers) in the figure. Below are two figures, both having the **same pixel size**: 432x432 but different dpi and figsize values:

6x6 * 72 dpi = 432x432 pixel size

3x3 * 144 dpi = 432x432 pixel size

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

#creates a new figure with a custom size of 9x10 inches.
plt.figure(figsize=(6, 6), dpi=72)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Plot with specified marker, linestyle, and color
plt.plot(x, y, marker='D', linestyle='dotted', color="blue")

# Enabling X-axis grid lines:
plt.grid(color="red")

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])

#creates a new figure with a custom size of 9x10 inches.
plt.figure(figsize=(3, 3), dpi=144)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

# Plot with specified marker, linestyle, and color
plt.plot(x, y, marker='D', linestyle='dotted', color="blue")

# Enabling X-axis grid lines:
plt.grid(color="red")

plt.show()
```

As you can see in the above images, the figure with a larger dpi has wider lines, axes, a larger font, and fewer tick labels - dpi magnifies all elements.
