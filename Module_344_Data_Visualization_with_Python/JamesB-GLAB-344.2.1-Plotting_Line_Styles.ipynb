## **Example 1: Dashdot Line style**


### We can use **'dashdot'** value in Linestyle argument.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, linestyle = "dashdot")
plt.show()
```

## **Example 2: Dotted Line style**

### We can use **':'** or **dotted** value in Linestyle argument.

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, linestyle = ":")
plt.show()
```

## **Example 3: Dash Line style**

### We can use **'--'** or  **dashed** value in Linestyle argument.

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, linestyle = "--")
plt.show()
```

## **Example 4: Dash Solid Line style**

### We can use **'-'** or **'solid'** value in Linestyle argument.
- use the `linestyle` keyword argument, or shorter `ls` to change the style of the plotted line:

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, ls = "-")
plt.show()
```

```python
# Using negative integers

xpoints_2 = np.array([3, 6])
ypoints_2 = np.array([2, 4])

plt.plot(xpoints_2, ypoints_2, linestyle= "-")
plt.show()
```
